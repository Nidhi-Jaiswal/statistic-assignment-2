1. What is hypothesis testing in statistics?

 --> Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data. It involves evaluating two competing hypotheses about a population parameter to determine which is more supported by the data.
 #### Steps in Hypothesis Testing:
 State the null and alternative hypotheses.

 Choose a significance level (Œ±).

 Select an appropriate test statistic and compute its value from the sample data.

 Determine the critical value or p-value.

 Compare the p-value to Œ± or the test statistic to the critical value.

 Make a decision (reject or fail to reject H‚ÇÄ).

 Draw a conclusion in the context of the problem.
 
 #### Errors in Hypothesis Testing:
 Type I Error (Œ±): Rejecting the null hypothesis when it is true.

 Type II Error (Œ≤): Failing to reject the null hypothesis when it is false.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

 --> The null hypothesis (H‚ÇÄ) and the alternative hypothesis (H‚ÇÅ or Ha) are two competing statements in hypothesis testing that represent different claims about a population parameter. They are central to statistical inference and help determine whether observed data provides enough evidence to support a specific claim.
 ### Null Hypothesis (H‚ÇÄ):
 The null hypothesis is a statement of no effect, no difference, or no relationship. It represents the default or status quo       assumption that there is nothing unusual or significant happening.

 It is the hypothesis that is initially assumed to be true unless the data provides strong evidence against it.

 It is often formulated with an equality sign(e.g., =, ‚â§, ‚â•).
 ##### Example:

            The mean height of a population is 170 cm" (H‚ÇÄ: Œº = 170).
            
            The new drug has no effect on blood pressure" (H‚ÇÄ: Œº‚ÇÅ - Œº‚ÇÇ = 0).
 ### Alternative Hypothesis (H‚ÇÅ or Ha):
 The alternative hypothesis is a statement that contradicts the null hypothesis. It represents the effect, difference, or         relationship that the researcher is interested in testing.

 It is what you hope to provide evidence for through the data.

 It is often formulated with an inequality sign (e.g., ‚â†, >, <).

 ##### Example:

            The mean height of a population is not 170 cm" (H‚ÇÅ: Œº ‚â† 170).

            The new drug reduces blood pressure" (H‚ÇÅ: Œº‚ÇÅ - Œº‚ÇÇ < 0).
 ### Relationship Between H‚ÇÄ and H‚ÇÅ:
 The null and alternative hypotheses are mutually exclusive and exhaustive. This means that if one is true, the other must be     false, and there are no other possibilities.

 Hypothesis testing is designed to determine whether there is enough evidence in the sample data to reject the null hypothesis     in favor of the alternative hypothesis.

3. What is the significance level in hypothesis testing, and why is it important?

 --> The significance level in hypothesis testing, denoted by Œ± (alpha), is a threshold that determines the probability of rejecting the null hypothesis (H‚ÇÄ) when it is actually true. It is a critical concept in statistical inference because it helps control the risk of making incorrect decisions based on sample data.
 ### Definition of Significance Level (Œ±):
 The significance level is the maximum probability of making a Type I error (rejecting the null hypothesis when it is true).

 It is set by the researcher before conducting the test and is often chosen as 0.05 (5%), 0.01 (1%), or 0.10 (10%), depending on   the context and the desired level of confidence.
 ### Why is the Significance Level Important?
 ##### Controls Type I Error:
 The significance level directly controls the probability of making a Type I error (false positive). For example, if Œ± = 0.05,     there is a 5% chance of incorrectly rejecting the null hypothesis when it is true.

 ##### Determines the Threshold for Decision-Making:
 The significance level is used to compare the p-value (the probability of observing the test statistic or something more         extreme, assuming H‚ÇÄ is true). If the p-value ‚â§ Œ±, the null hypothesis is rejected.

 ##### Balances Rigor and Flexibility:
 A lower significance level (e.g., Œ± = 0.01) makes the test more stringent, reducing the chance of Type I error but increasing     the risk of Type II error (failing to reject H‚ÇÄ when it is false). A higher significance level (e.g., Œ± = 0.10) makes the test   less stringent, increasing the chance of Type I error but reducing the risk of Type II error.
 ##### Standardizes Hypothesis Testing:
 By setting a significance level, researchers ensure consistency and objectivity in interpreting results across different         studies.



4. What does a P-value represent in hypothesis testing?

 --> The p-value is one of the most important concepts in hypothesis testing. It quantifies the strength of the evidence against the null hypothesis (H‚ÇÄ) and helps determine whether to reject or fail to reject H‚ÇÄ. Here's a detailed explanation of what a p-value represents:
 ### Definition of P-value:
The p-value is the probability of observing the test statistic or something more extreme, assuming that the null hypothesis (H‚ÇÄ) is true. It measures how consistent the observed data is with the null hypothesis.
 ### How is the P-value Calculated?
 The p-value is calculated based on the test statistic (e.g., t-statistic, z-score, chi-square statistic) and the distribution     of the test statistic under the null hypothesis. The steps are as follows:

 Compute the test statistic using the sample data.

 Determine the probability of observing a test statistic as extreme as, or more extreme than, the one calculated, assuming H‚ÇÄ is    true.

 This probability is the p-value.


5. How do you interpret the P-value in hypothesis testing?

 --> Interpreting the p-value in hypothesis testing is crucial for making informed decisions about the null hypothesis (H‚ÇÄ). Here's a step-by-step guide to understanding and interpreting the p-value:
 ### Steps to Interpret the P-value:
 ##### Compare the P-value to the Significance Level (Œ±):
 The significance level (Œ±) is a pre-defined threshold (e.g., 0.05, 0.01) that determines the cutoff for rejecting H‚ÇÄ.

 If the p-value ‚â§ Œ±, reject the null hypothesis (H‚ÇÄ). This indicates that the observed data is unlikely under H‚ÇÄ, providing       evidence for the alternative hypothesis (H‚ÇÅ).

 If the p-value > Œ±, fail to reject the null hypothesis (H‚ÇÄ). This indicates that the observed data is consistent with H‚ÇÄ, and     there is not enough evidence to support H‚ÇÅ.

 ##### Understand the Strength of Evidence:
 A small p-value (e.g., p < 0.05) suggests strong evidence against H‚ÇÄ, indicating that the observed data is unlikely to occur if   H‚ÇÄ were true.
 
 A large p-value (e.g., p > 0.05) suggests weak evidence against H‚ÇÄ, indicating that the observed data is consistent with H‚ÇÄ.

 ##### Avoid Misinterpretations:
 The p-value is not the probability that H‚ÇÄ is true or false. It is the probability of the data (or something more extreme)       given that H‚ÇÄ is true.

 The p-value does not measure the size or importance of an effect. A small p-value does not necessarily mean the effect is large   or practically significant.
 ### Examples of P-value Interpretation:
 ##### Example 1: Small P-value (p ‚â§ Œ±)
 Scenario: A researcher tests whether a new drug lowers blood pressure. They set Œ± = 0.05 and calculate a p-value of 0.01.

 Interpretation:

            The p-value (0.01) is less than Œ± (0.05), so the null hypothesis (H‚ÇÄ: the drug has no effect) is rejected.

            There is strong evidence to support the alternative hypothesis (H‚ÇÅ: the drug lowers blood pressure).
 ##### Example 2: Large P-value (p > Œ±)
 Scenario: A researcher tests whether a coin is fair (H‚ÇÄ: probability of heads = 0.5). They set Œ± = 0.05 and calculate a p-value   of 0.30.

 Interpretation:

            The p-value (0.30) is greater than Œ± (0.05), so the null hypothesis (H‚ÇÄ: the coin is fair) is not rejected.

            There is not enough evidence to conclude that the coin is unfair.

6. What are Type 1 and Type 2 errors in hypothesis testing?

 --> In hypothesis testing, Type I and Type II errors are two types of mistakes that can occur when making decisions about the null hypothesis (H‚ÇÄ). Understanding these errors is crucial because they represent the trade-offs between being too strict or too lenient in rejecting or failing to reject H‚ÇÄ.
 ### Type I Error (False Positive):
 ##### Definition: 
 A Type I error occurs when the null hypothesis (H‚ÇÄ) is rejected when it is actually true.

 ##### Probability: 
 The probability of making a Type I error is denoted by Œ± (alpha), which is the significance level of the test.

 ##### Example:
 Suppose a medical test is designed to detect a disease. A Type I error occurs if the test indicates that a healthy person has     the disease (false positive).

 ##### Consequence:
 Type I errors can lead to unnecessary actions or interventions, such as prescribing medication when it is not needed.
 ### Type II Error (False Negative):
 ##### Definition:
 A Type II error occurs when the null hypothesis (H‚ÇÄ) is not rejected when it is actually false.

 ##### Probability: 
 The probability of making a Type II error is denoted by Œ≤ (beta).

 ##### Example:
 In the same medical test, a Type II error occurs if the test fails to detect the disease in a person who actually has it (false   negative).

 ##### Consequence:
 Type II errors can lead to missed opportunities for intervention, such as failing to treat a disease that is present.
 ### Relationship Between Type I and Type II Errors:
 There is an inverse relationship between Type I and Type II errors. Reducing the probability of one type of error typically       increases the probability of the other.

 ##### For example:

          Lowering the significance level (Œ±) reduces the chance of Type I error but increases the chance of Type II error.

          Increasing the sample size can reduce both Type I and Type II errors.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

 --> In hypothesis testing, the choice between a one-tailed test and a two-tailed test depends on the research question and the directionality of the hypothesis being tested. The key difference lies in how the alternative hypothesis (H‚ÇÅ) is formulated and how the critical region is defined.
 ### One-Tailed Test:
 ##### Definition: 
 A one-tailed test is used when the research hypothesis specifies a direction of the effect or difference. It tests whether a     parameter is either greater than or less than a certain value, but not both.

 ##### Alternative Hypothesis (H‚ÇÅ):
 H‚ÇÅ is directional (e.g., Œº > Œº‚ÇÄ or Œº < Œº‚ÇÄ).

 ##### Critical Region:
 The critical region is located entirely in one tail of the distribution (either the right tail or the left tail).

 ##### When to Use: 
 Use a one-tailed test when you are only interested in detecting an effect in one direction.

 ##### Example:
 Testing whether a new drug increases the average recovery rate (H‚ÇÅ: Œº > Œº‚ÇÄ).
 ###  Two-Tailed Test:
 ##### Definition: 
 A two-tailed test is used when the research hypothesis does not specify a direction of the effect or difference. It tests         whether a parameter is not equal to a certain value, meaning it could be either greater than or less than.

 ##### Alternative Hypothesis (H‚ÇÅ):
 H‚ÇÅ is non-directional (e.g., Œº ‚â† Œº‚ÇÄ).

 ##### Critical Region:
 The critical region is split between both tails of the distribution.

 ##### When to Use:
 Use a two-tailed test when you are interested in detecting any difference, regardless of direction.

 ##### Example:
 Testing whether a new drug has any effect on the average recovery rate (H‚ÇÅ: Œº ‚â† Œº‚ÇÄ).

20. What is an ANOVA test, and what are its assumptions?

 --> The ANOVA (Analysis of Variance) test is a statistical method used to compare the means of three or more groups to determine if there is any significant difference among them. ANOVA helps test the hypothesis that the means of the groups are equal by comparing the variance between the groups to the variance within each group.
 ### Why Use ANOVA?
 Instead of performing multiple t-tests to compare pairs of groups, ANOVA allows for a more efficient comparison across multiple    groups simultaneously, controlling for the risk of Type I errors (false positives).
 ### Assumptions of ANOVA
For ANOVA results to be valid, certain assumptions must be met:

 ##### Independence of Observations
 Each sample must be drawn independently. This means that the measurements or data points from one group should not influence or   be related to those from another group.

 Example: The data points from one treatment group (e.g., a drug group) should not affect the data from another group (e.g., a     placebo group).
 ##### Normality
 The data in each group should follow a normal distribution. While ANOVA is somewhat robust to deviations from normality, severe   deviations can lead to inaccurate conclusions.

 Shapiro-Wilk test or Q-Q plots can be used to check normality.
 ##### Homogeneity of Variances (Homoscedasticity)
 The variance within each group should be roughly equal. This assumption ensures that the spread of data is similar across all     groups.

 Levene's test or Bartlett's test can be used to test for equal variances.
 ##### Random Sampling
 The data should be randomly selected from the population to ensure that the sample is representative of the broader population.

21. What are the different types of ANOVA tests?

 --> There are several types of ANOVA tests used to analyze the variance in different experimental designs. The main types include One-Way ANOVA, Two-Way ANOVA, and their variations (e.g., Repeated Measures ANOVA, Multivariate ANOVA (MANOVA)). Here‚Äôs an overview of the different types of ANOVA tests:
 ### One-Way ANOVA
 One-Way ANOVA is used when there is one independent variable with three or more levels (groups) and you want to compare their     means. It tests if there is a significant difference in the means of the groups based on a single factor.

 ##### Example Use Case:
 Comparing the test scores of students from three different teaching methods (Method A, Method B, Method C).
 ##### Hypotheses:
        ùêª0: All group means are equal.
      
        ùêªùëé: At least one group mean is different.
 ### Two-Way ANOVA
 Two-Way ANOVA is used when there are two independent variables and you want to test the effect of each on the dependent           variable, both individually and interactively. It allows for the analysis of interactions between two factors.

 ##### Example Use Case:
 Analyzing how teaching method (Method A, Method B) and study time (Short, Medium, Long) affect students' test scores.
 ##### Hypotheses:
        ùêª0: The main effects of both factors (teaching method and study time) and their interaction have no effect on the                        dependent variable.
        
        ùêªùëé: At least one of the factors or the interaction effect significantly affects the dependent variable.
 Two-Way ANOVA can also be performed with replication (multiple observations within each combination of factors), which allows     you to assess the main effects and interactions while accounting for variability within the groups.
 ### Repeated Measures ANOVA
 Repeated Measures ANOVA is used when the same subjects are measured multiple times under different conditions or at different     time points. It compares means when there is within-subject correlation and helps control for variability between subjects.

 ##### Example Use Case:
 Measuring blood pressure at three different time points (before treatment, after 1 month, and after 3 months) in the same group   of patients.
 ##### Hypotheses:
        ùêª0: There is no significant difference in means across the repeated measures.
        
        ùêªùëé: At least one mean is significantly different from the others.
 ### Multivariate Analysis of Variance (MANOVA)
 MANOVA is an extension of ANOVA that allows you to test the differences in multiple dependent variables simultaneously. This is   useful when you have multiple response variables that are correlated.

 ##### Example Use Case:
 Investigating the effects of diet on both weight loss and cholesterol levels in a group of people.
 ##### Hypotheses:
       ùêª0: There is no significant difference in the means of the dependent variables across the groups.
       
       ùêªùëé: There is a significant difference in at least one of the dependent variables.
 ### Mixed-Design ANOVA
 Mixed-Design ANOVA is used when you have both between-subjects factors (different groups) and within-subjects factors (repeated   measurements). This design is used when some factors are manipulated between groups and others within the same group.

 ##### Example Use Case:
 Studying the impact of diet type (A, B, C) on weight loss over three different time periods (1 month, 3 months, and 6 months),   where the diet type is a between-subject factor and time is a within-subject factor.
 ### One-Way ANCOVA (Analysis of Covariance)
 ANCOVA is used when you want to compare the means of a dependent variable across multiple groups while controlling for one or     more covariates (continuous variables) that may influence the dependent variable.

 ##### Example Use Case:
 Comparing the effectiveness of three diets on weight loss while controlling for the effect of age and gender.

22. What is the F-test, and how does it relate to hypothesis testing?

 --> The F-test is a statistical test used to compare two variances and assess whether they are significantly different. It is commonly used in the context of Analysis of Variance (ANOVA), regression analysis, and to test for the equality of variances in different groups.

The F-test works by comparing the ratio of two variances, as follows:
                        
                        F = Variance¬†Between¬†Groups / Variance¬†Within¬†Groups
 ### Key Characteristics of the F-Test:
 It involves the F-distribution, which is right-skewed and non-negative.
 
 The degrees of freedom for the numerator (variance between groups) and denominator (variance within groups) are required to       calculate the critical value of the F-distribution.
 
 The F-statistic is used to assess if the group means (in the case of ANOVA) or the regression model (in the case of regression   analysis) are significantly different.
 ### F-Test in Hypothesis Testing: Steps Summary
 ##### State the Hypotheses:

 For variances: ùêª0: Equal variances, ùêªùëé: Unequal variances.

 For ANOVA: H0: Equal means, ùêªùëé: At least one mean is different.

 For regression: ùêª0: No significant model, ùêªùëé: The model explains significant variance.
 ##### Calculate the F-statistic:

 For variances:

                     F = Variance¬†1 / Variance¬†2
 For ANOVA: 
 
                     F = Variance¬†Between¬†Groups / Variance¬†Within¬†Groups
 For regression: 
 
                     F = Explained¬†Variance / Unexplained¬†Variance
 ##### Compare the F-statistic to the Critical Value:
 Use the F-distribution table with the appropriate degrees of freedom to determine the critical value for your chosen             significance level (ùõº, typically 0.05).

 ##### Make a Decision:
  If the F-statistic is greater than the critical value, reject ùêª0.
  
  If the F-statistic is less than or equal to the critical value, fail to reject ùêª0.

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
import scipy.stats as stats

# Function to perform Z-test
def z_test(sample, pop_mean, pop_std, alpha=0.05):
    
    # Sample statistics
    sample_mean = np.mean(sample)  # Calculate sample mean
    sample_size = len(sample)      # Sample size

    # Compute Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # cdf gives area to the left

    # Interpretation of results
    print(f"Sample Mean: {sample_mean:.3f}")
    print(f"Z-Score: {z_score:.3f}")
    print(f"P-Value: {p_value:.5f}")

    if p_value < alpha:
        print("Result: Reject the Null Hypothesis (Significant difference detected)")
    else:
        print("Result: Fail to Reject the Null Hypothesis (No significant difference detected)")

# Example Usage
sample_data = [52, 48, 50, 53, 49, 51, 47, 50, 54, 49]  # Sample data
population_mean = 50  # Known population mean
population_std = 3    # Known population standard deviation

# Perform Z-test
z_test(sample_data, population_mean, population_std)
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to perform Two-Tailed Z-Test
def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Performs a two-tailed Z-test and visualizes the decision region.

    Parameters:
    - sample: list or numpy array (Sample data)
    - pop_mean: float (Population mean)
    - pop_std: float (Population standard deviation)
    - alpha: float (Significance level, default is 0.05)

    Returns:
    - Z-score, p-value, and graphical representation of the hypothesis test
    """

    # Sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Compute Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Compute critical Z-value for two-tailed test
    z_critical = stats.norm.ppf(1 - alpha/2)

    # Print results
    print(f"Sample Mean: {sample_mean:.3f}")
    print(f"Z-Score: {z_score:.3f}")
    print(f"P-Value: {p_value:.5f}")
    print(f"Critical Z-Value: ¬±{z_critical:.3f}")

    # Decision rule
    if abs(z_score) > z_critical:
        print("Result: Reject the Null Hypothesis (Significant difference detected)")
    else:
        print("Result: Fail to Reject the Null Hypothesis (No significant difference detected)")

    # Visualization
    visualize_decision_region(z_score, z_critical)

# Function to visualize the decision region
def visualize_decision_region(z_score, z_critical):
    """
    Plots the normal distribution with rejection regions highlighted.
    """
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label="Standard Normal Distribution", color='black')

    # Fill the rejection regions
    plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.5, label="Rejection Region")

    # Draw the critical values
    plt.axvline(-z_critical, linestyle="--", color="red", label=f"Critical Value (-{z_critical:.2f})")
    plt.axvline(z_critical, linestyle="--", color="red", label=f"Critical Value (+{z_critical:.2f})")

    # Draw the computed Z-score
    plt.axvline(z_score, linestyle="--", color="blue", label=f"Computed Z-Score ({z_score:.2f})")

    # Labels and legend
    plt.title("Two-Tailed Z-Test: Decision Region")
    plt.xlabel("Z-Score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

# Example Usage
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulated sample data
population_mean = 100  # Known population mean
population_std = 15     # Known population standard deviation

# Perform Two-Tailed Z-Test
two_tailed_z_test(sample_data, population_mean, population_std)
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type1_type2_errors(mu_null=100, mu_alt=105, sigma=15, n=30, alpha=0.05):
    """
    Visualizes Type 1 and Type 2 errors in hypothesis testing.

    Parameters:
    - mu_null: Population mean under H0 (default: 100)
    - mu_alt: Population mean under H1 (default: 105)
    - sigma: Population standard deviation (default: 15)
    - n: Sample size (default: 30)
    - alpha: Significance level (default: 0.05)
    """

    # Compute standard error
    se = sigma / np.sqrt(n)

    # Compute critical value for rejection region (two-tailed test)
    z_critical = stats.norm.ppf(1 - alpha / 2)
    critical_value_right = mu_null + z_critical * se
    critical_value_left = mu_null - z_critical * se

    # Compute Type 2 error (Œ≤)
    beta_left = stats.norm.cdf(critical_value_left, loc=mu_alt, scale=se)
    beta_right = 1 - stats.norm.cdf(critical_value_right, loc=mu_alt, scale=se)
    beta = beta_left + beta_right

    # Compute power (1 - Œ≤)
    power = 1 - beta

    # Define x range for plotting
    x = np.linspace(mu_null - 4 * se, mu_alt + 4 * se, 1000)

    # Compute normal distributions for H0 and H1
    y_null = stats.norm.pdf(x, loc=mu_null, scale=se)
    y_alt = stats.norm.pdf(x, loc=mu_alt, scale=se)

    # Plot distributions
    plt.figure(figsize=(10, 5))
    plt.plot(x, y_null, label="Null Hypothesis (H0)", color='blue')
    plt.plot(x, y_alt, label="Alternative Hypothesis (H1)", color='green')

    # Shade Type 1 Error (Œ±) region
    plt.fill_between(x, y_null, where=(x >= critical_value_right) | (x <= critical_value_left),
                     color='red', alpha=0.3, label="Type 1 Error (Œ±)")

    # Shade Type 2 Error (Œ≤) region
    plt.fill_between(x, y_alt, where=(x > critical_value_left) & (x < critical_value_right),
                     color='orange', alpha=0.3, label="Type 2 Error (Œ≤)")

    # Labels and legend
    plt.axvline(critical_value_left, linestyle="--", color="black", label=f"Critical Value ({critical_value_left:.2f})")
    plt.axvline(critical_value_right, linestyle="--", color="black", label=f"Critical Value ({critical_value_right:.2f})")
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

    # Print results
    print(f"Critical Values: Left = {critical_value_left:.2f}, Right = {critical_value_right:.2f}")
    print(f"Type 1 Error (Œ±): {alpha:.3f}")
    print(f"Type 2 Error (Œ≤): {beta:.3f}")
    print(f"Power of the Test (1 - Œ≤): {power:.3f}")

# Example Usage
visualize_type1_type2_errors()
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def independent_t_test(group1, group2, alpha=0.05):
    """
    Performs an independent (two-sample) T-test and interprets the results.

    Parameters:
    - group1: List or NumPy array (Sample data for group 1)
    - group2: List or NumPy array (Sample data for group 2)
    - alpha: Significance level (default: 0.05)

    Returns:
    - T-statistic, p-value, and conclusion about rejecting/failing to reject H0.
    """

    # Compute sample statistics
    mean1, mean2 = np.mean(group1), np.mean(group2)
    std1, std2 = np.std(group1, ddof=1), np.std(group2, ddof=1)
    n1, n2 = len(group1), len(group2)

    # Test for equal variances using Levene's test
    stat, p_var = stats.levene(group1, group2)
    equal_var = p_var > alpha  # If p > alpha, assume equal variances

    # Perform independent T-test
    t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=equal_var)

    # Print results
    print(f"Group 1 Mean: {mean1:.2f}, Std Dev: {std1:.2f}, Sample Size: {n1}")
    print(f"Group 2 Mean: {mean2:.2f}, Std Dev: {std2:.2f}, Sample Size: {n2}")
    print(f"Levene‚Äôs Test for Equal Variance: p-value = {p_var:.4f} (Equal Variance Assumed: {equal_var})")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.5f}")

    # Decision rule
    if p_value < alpha:
        print("Result: Reject the Null Hypothesis (Significant difference between groups)")
    else:
        print("Result: Fail to Reject the Null Hypothesis (No significant difference between groups)")

    # Visualization
    visualize_t_test(group1, group2)

def visualize_t_test(group1, group2):
    """
    Plots histograms of both groups with KDE (Kernel Density Estimation).
    """
    plt.figure(figsize=(10, 5))
    sns.histplot(group1, kde=True, color='blue', label="Group 1", alpha=0.5)
    sns.histplot(group2, kde=True, color='green', label="Group 2", alpha=0.5)

    plt.axvline(np.mean(group1), color='blue', linestyle="--", label="Mean Group 1")
    plt.axvline(np.mean(group2), color='green', linestyle="--", label="Mean Group 2")

    plt.title("Independent T-Test: Group Distributions")
    plt.xlabel("Values")
    plt.ylabel("Density")
    plt.legend()
    plt.grid()
    plt.show()

# Example Usage: Simulating Two Independent Samples
np.random.seed(42)
group1 = np.random.normal(loc=100, scale=15, size=30)  # Sample 1: Mean=100, Std=15, Size=30
group2 = np.random.normal(loc=105, scale=15, size=30)  # Sample 2: Mean=105, Std=15, Size=30

# Perform Independent T-Test
independent_t_test(group1, group2)
```

```python
#7. Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def paired_t_test(before, after, alpha=0.05):
    """
    Performs a paired sample T-test and visualizes the results.

    Parameters:
    - before: List or NumPy array (Measurements before intervention)
    - after: List or NumPy array (Measurements after intervention)
    - alpha: Significance level (default: 0.05)
    """

    # Compute the differences
    differences = np.array(after) - np.array(before)
    
    # Perform paired T-test
    t_stat, p_value = stats.ttest_rel(before, after)

    # Print results
    print(f"Before Mean: {np.mean(before):.2f}, Std Dev: {np.std(before, ddof=1):.2f}")
    print(f"After Mean: {np.mean(after):.2f}, Std Dev: {np.std(after, ddof=1):.2f}")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.5f}")

    # Decision rule
    if p_value < alpha:
        print("Result: Reject the Null Hypothesis (Significant difference between before and after)")
    else:
        print("Result: Fail to Reject the Null Hypothesis (No significant difference)")

    # Visualization
    visualize_paired_t_test(before, after, differences)

def visualize_paired_t_test(before, after, differences):
    """
    Plots before-after values and a distribution of differences.
    """
    plt.figure(figsize=(12, 5))

    # 1. Before-After Paired Line Plot
    plt.subplot(1, 2, 1)
    for i in range(len(before)):
        plt.plot([1, 2], [before[i], after[i]], marker='o', color='gray', alpha=0.6)
    plt.scatter([1]*len(before), before, color='blue', label="Before", alpha=0.7)
    plt.scatter([2]*len(after), after, color='green', label="After", alpha=0.7)
    plt.xticks([1, 2], ["Before", "After"])
    plt.title("Before vs After (Paired Data)")
    plt.ylabel("Values")
    plt.grid()
    plt.legend()

    # 2. Distribution of Differences
    plt.subplot(1, 2, 2)
    sns.histplot(differences, kde=True, bins=10, color="purple", alpha=0.6)
    plt.axvline(np.mean(differences), color="red", linestyle="--", label="Mean Difference")
    plt.title("Distribution of Differences (After - Before)")
    plt.xlabel("Difference")
    plt.ylabel("Density")
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()

# Example Usage: Simulating Before and After Data
np.random.seed(42)
before = np.random.normal(loc=100, scale=10, size=30)  # Measurements before intervention
after = before + np.random.normal(loc=2, scale=5, size=30)  # After treatment (mean improvement of 2)

# Perform Paired T-Test
paired_t_test(before, after)
```

```python
#8. Simulate data and perform both Z-test and T-test, then compare the results using Python.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def perform_z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Performs a one-sample Z-test and interprets the results.
    """
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    
    # Compute Z-statistic
    se = pop_std / np.sqrt(sample_size)  # Standard error
    z_stat = (sample_mean - pop_mean) / se
    
    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
    
    # Print results
    print("\n===== Z-Test Results =====")
    print(f"Sample Mean: {sample_mean:.2f}, Population Mean: {pop_mean}")
    print(f"Z-Statistic: {z_stat:.4f}, P-Value: {p_value:.5f}")

    if p_value < alpha:
        print("Result: Reject the Null Hypothesis (Significant difference)")
    else:
        print("Result: Fail to Reject the Null Hypothesis (No significant difference)")
    
    return z_stat, p_value

def perform_t_test(sample, pop_mean, alpha=0.05):
    """
    Performs a one-sample T-test and interprets the results.
    """
    t_stat, p_value = stats.ttest_1samp(sample, pop_mean)

    # Print results
    print("\n===== T-Test Results =====")
    print(f"T-Statistic: {t_stat:.4f}, P-Value: {p_value:.5f}")

    if p_value < alpha:
        print("Result: Reject the Null Hypothesis (Significant difference)")
    else:
        print("Result: Fail to Reject the Null Hypothesis (No significant difference)")
    
    return t_stat, p_value

def visualize_tests(sample, pop_mean, pop_std):
    """
    Visualizes the sample distribution, population mean, and confidence intervals.
    """
    plt.figure(figsize=(10, 5))
    
    # Plot sample distribution
    sns.histplot(sample, kde=True, bins=15, color="blue", alpha=0.5, label="Sample Distribution")
    
    # Mark population mean
    plt.axvline(pop_mean, color="red", linestyle="--", label="Population Mean")
    
    # Mark sample mean
    plt.axvline(np.mean(sample), color="green", linestyle="--", label="Sample Mean")
    
    # Confidence Interval (95%)
    sample_size = len(sample)
    se = pop_std / np.sqrt(sample_size)
    ci_lower = np.mean(sample) - 1.96 * se
    ci_upper = np.mean(sample) + 1.96 * se
    plt.axvline(ci_lower, color="purple", linestyle="--", label="95% CI Lower")
    plt.axvline(ci_upper, color="purple", linestyle="--", label="95% CI Upper")

    plt.title("Sample Distribution vs. Population Mean")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid()
    plt.show()

# === Simulating Data ===
np.random.seed(42)

# Population parameters
pop_mean = 100  # Known population mean
pop_std = 15    # Known population standard deviation

# Generate a random sample from normal distribution
sample_size = 30
sample = np.random.normal(loc=pop_mean + 2, scale=pop_std, size=sample_size)  # Shifted mean

# === Perform Tests ===
z_stat, z_p_value = perform_z_test(sample, pop_mean, pop_std)
t_stat, t_p_value = perform_t_test(sample, pop_mean)

# === Visualize the Distribution ===
visualize_tests(sample, pop_mean, pop_std)
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95, pop_std=None):
    """
    Calculates the confidence interval for a sample mean.

    Parameters:
    - sample: List or NumPy array (sample data)
    - confidence: Confidence level (default: 95%)
    - pop_std: Population standard deviation (if known, use Z-distribution)

    Returns:
    - Tuple (lower_bound, upper_bound) representing the confidence interval
    """

    sample_mean = np.mean(sample)
    sample_size = len(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (unbiased estimate)

    # Determine the appropriate critical value
    if pop_std is not None:
        # Use Z-distribution if population standard deviation is known
        critical_value = stats.norm.ppf(1 - (1 - confidence) / 2)
        standard_error = pop_std / np.sqrt(sample_size)
    else:
        # Use T-distribution if population standard deviation is unknown
        critical_value = stats.t.ppf(1 - (1 - confidence) / 2, df=sample_size - 1)
        standard_error = sample_std / np.sqrt(sample_size)

    # Compute the confidence interval
    margin_of_error = critical_value * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# === Example Usage ===
np.random.seed(42)

# Simulating a sample dataset
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Mean=100, Std=15, Size=30

# 95% Confidence Interval (Unknown Population Std ‚Üí Uses T-distribution)
ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence=0.95, pop_std=None):
    """
    Calculates the margin of error for a given confidence level.

    Parameters:
    - sample: List or NumPy array (sample data)
    - confidence: Confidence level (default: 95%)
    - pop_std: Population standard deviation (if known, use Z-distribution)

    Returns:
    - Margin of error (MoE)
    """
    sample_size = len(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (unbiased estimate)

    # Determine the appropriate critical value
    if pop_std is not None:
        # Use Z-distribution if population standard deviation is known
        critical_value = stats.norm.ppf(1 - (1 - confidence) / 2)
        standard_error = pop_std / np.sqrt(sample_size)
    else:
        # Use T-distribution if population standard deviation is unknown
        critical_value = stats.t.ppf(1 - (1 - confidence) / 2, df=sample_size - 1)
        standard_error = sample_std / np.sqrt(sample_size)

    # Compute the margin of error
    moe = critical_value * standard_error
    return moe

# === Example Usage ===
np.random.seed(42)

# Simulating a sample dataset
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Mean=100, Std=15, Size=30

# 95% Margin of Error (Unknown Population Std ‚Üí Uses T-distribution)
moe_95 = margin_of_error(sample_data, confidence=0.95)
print(f"95% Margin of Error: {moe_95:.2f}")

# 99% Margin of Error (Higher Confidence Level)
moe_99 = margin_of_error(sample_data, confidence=0.99)
print(f"99% Margin of Error: {moe_99:.2f}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayesian_inference(prior, likelihood, evidence):
    """
    Performs Bayesian Inference using Bayes' Theorem.

    Parameters:
    - prior: Prior probability of hypothesis (P(H))
    - likelihood: Likelihood of evidence given hypothesis (P(E | H))
    - evidence: Marginal likelihood of evidence (P(E))

    Returns:
    - Posterior probability (P(H | E))
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# === Example Usage ===
# Prior Probability: P(H) - Prior belief of having the disease (e.g., 1% chance)
prior_disease = 0.01

# Likelihood: P(E | H) - Probability of testing positive given that the person has the disease (e.g., 90% sensitivity)
likelihood_positive_given_disease = 0.9

# Evidence: P(E) - Probability of testing positive (e.g., both people with and without the disease)
# P(E) = P(E | H) * P(H) + P(E | ~H) * P(~H)
probability_positive = (likelihood_positive_given_disease * prior_disease) + (0.05 * (1 - prior_disease))

# Compute Posterior Probability: P(H | E) - The probability of having the disease given a positive test result
posterior_disease_given_positive = bayesian_inference(prior_disease, likelihood_positive_given_disease, probability_positive)

print(f"Posterior Probability of having the disease given a positive test result: {posterior_disease_given_positive:.4f}")
```

```python
#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test based on observed data.

    Parameters:
    - observed: A 2D NumPy array or list of lists representing the observed frequency table.

    Returns:
    - expected: A 2D NumPy array representing the expected frequency table.
    """
    # Calculate row totals
    row_totals = np.sum(observed, axis=1, keepdims=True)
    
    # Calculate column totals
    col_totals = np.sum(observed, axis=0, keepdims=True)
    
    # Calculate grand total (sum of all observed frequencies)
    grand_total = np.sum(observed)
    
    # Calculate expected frequencies using the formula: E_ij = (Row_Total_i * Col_Total_j) / Grand_Total
    expected = (row_totals * col_totals) / grand_total
    
    return expected

# === Example Usage ===
# Observed data: Contingency table for Gender vs Preference
# Columns: Like, Dislike
# Rows: Male, Female
observed_data = np.array([[30, 10],  # Male: 30 Like, 10 Dislike
                          [20, 20]]) # Female: 20 Like, 20 Dislike

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

print("Observed Frequencies:")
print(observed_data)
print("\nExpected Frequencies:")
print(expected_frequencies)
```

```python
#16. Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
import scipy.stats as stats

# Function to perform the F-test comparing the variances of two samples
def f_test(sample1, sample2):
    # Step 1: Calculate the sample variances
    var1 = np.var(sample1, ddof=1)  # ddof=1 for sample variance (N-1)
    var2 = np.var(sample2, ddof=1)

    # Step 2: Calculate the F-statistic (larger variance / smaller variance)
    f_statistic = var1 / var2 if var1 > var2 else var2 / var1

    # Step 3: Compute the degrees of freedom
    df1 = len(sample1) - 1  # Degrees of freedom for sample 1
    df2 = len(sample2) - 1  # Degrees of freedom for sample 2

    # Step 4: Calculate the p-value for the F-statistic
    p_value = 1 - stats.f.cdf(f_statistic, df1, df2)

    # Return the F-statistic, p-value, and variances
    return f_statistic, p_value, var1, var2

# Example Usage
# Generate two random samples
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Sample 1: mean=50, std=10, size=30
sample2 = np.random.normal(loc=50, scale=15, size=30)  # Sample 2: mean=50, std=15, size=30

# Perform the F-test
f_stat, p_val, var1, var2 = f_test(sample1, sample2)

# Display the results
print(f"Sample 1 Variance: {var1:.2f}")
print(f"Sample 2 Variance: {var2:.2f}")
print(f"F-statistic: {f_stat:.2f}")
print(f"P-value: {p_val:.4f}")

# Interpret the results
alpha = 0.05  # Significance level
if p_val < alpha:
    print("\nConclusion: Reject the null hypothesis - the variances are significantly different.")
else:
    print("\nConclusion: Fail to reject the null hypothesis - the variances are equal.")
```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Function to check assumptions for ANOVA
def check_anova_assumptions(groups):
    # 1. Check Normality of Each Group (Shapiro-Wilk test)
    normality_results = {}
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        normality_results[f'Group {i+1}'] = {'statistic': stat, 'p-value': p_value}
        
        # Plot Q-Q plot for normality
        plt.figure(figsize=(5, 4))
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f'Q-Q Plot for Group {i+1}')
        plt.show()

    # Print Normality Test Results
    print("Normality Test Results (Shapiro-Wilk Test):")
    for group, result in normality_results.items():
        print(f"{group}: Statistic = {result['statistic']:.4f}, p-value = {result['p-value']:.4f}")
    
    # 2. Check Homogeneity of Variances (Levene's Test)
    stat, p_value = stats.levene(*groups)
    print("\nLevene's Test for Homogeneity of Variances:")
    print(f"Statistic = {stat:.4f}, p-value = {p_value:.4f}")
    
    # Interpretation of Results
    alpha = 0.05
    if p_value < alpha:
        print("Conclusion: Variances are significantly different (violation of assumption).")
    else:
        print("Conclusion: Variances are equal (assumption met).")

    # 3. Visualize Data (Boxplot) for potential outliers
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=groups)
    plt.title('Boxplot of Groups')
    plt.xlabel('Groups')
    plt.ylabel('Values')
    plt.xticks([0, 1, 2, 3], [f'Group {i+1}' for i in range(len(groups))])
    plt.grid(True)
    plt.show()

# Example Usage: Comparing the means of 4 groups
np.random.seed(42)  # For reproducibility

# Generate random data for 4 groups with different means
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, n=30
group2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std=10, n=30
group3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60, Std=10, n=30
group4 = np.random.normal(loc=65, scale=10, size=30)  # Mean=65, Std=10, n=30

# List of groups
groups = [group1, group2, group3, group4]

# Check the assumptions for ANOVA
check_anova_assumptions(groups)
```

```python
#20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Simulating data for two-way ANOVA
np.random.seed(42)

# Factor 1: Diet Type (2 levels: High-carb, Low-carb)
# Factor 2: Gender (2 levels: Male, Female)

# Sample size per combination of factor levels
n = 30

# Generate data for each combination of diet and gender
diet_types = ['High-carb', 'Low-carb']
genders = ['Male', 'Female']

# Create a DataFrame to hold the data
data = []

for diet in diet_types:
    for gender in genders:
        # Simulate weight loss data for each group
        mean = 10 if diet == 'High-carb' and gender == 'Male' else 8
        std_dev = 3
        weight_loss = np.random.normal(mean, std_dev, n)
        
        # Append the data
        for w in weight_loss:
            data.append([diet, gender, w])

# Create the DataFrame
df = pd.DataFrame(data, columns=['Diet', 'Gender', 'Weight Loss'])

# Perform Two-Way ANOVA using the model
model = ols('Weight_Loss ~ C(Diet) + C(Gender) + C(Diet):C(Gender)', data=df).fit()

# Perform the ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

# Visualizing the results
# Boxplot for interaction between Diet and Gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='Diet', y='Weight Loss', hue='Gender', data=df)
plt.title('Two-Way ANOVA: Interaction Between Diet and Gender')
plt.xlabel('Diet Type')
plt.ylabel('Weight Loss')
plt.show()

# Interaction plot
sns.interactionplot(x=df['Diet'], y=df['Weight Loss'], trace=df['Gender'], markers=['D', 'o'], ms=10)
plt.title('Interaction Plot: Diet vs. Gender')
plt.xlabel('Diet Type')
plt.ylabel('Weight Loss')
plt.show()
```

```python
#21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Function to visualize the F-distribution
def visualize_f_distribution(dof1, dof2):
    # Generate a range of x values (F-statistic values)
    x = np.linspace(0, 5, 1000)
    
    # Calculate the PDF of the F-distribution
    y = stats.f.pdf(x, dfn=dof1, dfd=dof2)
    
    # Plot the F-distribution
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'F-distribution (dof1={dof1}, dof2={dof2})', color='blue')
    plt.fill_between(x, y, 0, alpha=0.3, color='blue')
    
    # Add labels and title
    plt.title(f'F-distribution (dof1={dof1}, dof2={dof2})', fontsize=14)
    plt.xlabel('F-statistic', fontsize=12)
    plt.ylabel('Density', fontsize=12)
    plt.legend(loc='upper right')
    
    # Display the plot
    plt.grid(True)
    plt.show()

# Example: Visualize F-distribution with dof1=5, dof2=20
visualize_f_distribution(dof1=5, dof2=20)
```

```python
#22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.

import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Simulating data for three groups
np.random.seed(42)

# Group 1 (e.g., Low-Carb diet)
group1 = np.random.normal(loc=5, scale=2, size=30)

# Group 2 (e.g., Medium-Carb diet)
group2 = np.random.normal(loc=7, scale=2.5, size=30)

# Group 3 (e.g., High-Carb diet)
group3 = np.random.normal(loc=6, scale=2, size=30)

# Combine the data into a DataFrame
data = pd.DataFrame({
    'Group': ['Low-Carb']*30 + ['Medium-Carb']*30 + ['High-Carb']*30,
    'Weight_Loss': np.concatenate([group1, group2, group3])
})

# Perform the one-way ANOVA test
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Print the results
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Boxplot to visualize the results
plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='Weight_Loss', data=data)
plt.title('One-Way ANOVA: Comparison of Weight Loss Across Diet Types')
plt.xlabel('Diet Type')
plt.ylabel('Weight Loss')
plt.show()

# Interpretation
if p_value < 0.05:
    print("\nSince the p-value is less than 0.05, we reject the null hypothesis.")
    print("There is a significant difference in the mean weight loss across the different diet types.")
else:
    print("\nSince the p-value is greater than 0.05, we fail to reject the null hypothesis.")
    print("There is no significant difference in the mean weight loss across the different diet types.")
```

```python
#23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.

import numpy as np
import scipy.stats as stats

# Step 1: Simulate random data from a normal distribution
np.random.seed(42)

# Parameters for the normal distribution
population_mean = 50  # Known population mean
population_std = 10   # Standard deviation
sample_size = 100     # Sample size

# Simulate the sample data (e.g., sample from a population)
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Step 2: Perform a one-sample t-test
# Hypothesis:
# H‚ÇÄ: The sample mean is equal to the population mean
# H‚ÇÅ: The sample mean is not equal to the population mean (two-tailed test)
sample_mean = np.mean(sample_data)
t_statistic, p_value = stats.ttest_1samp(sample_data, population_mean)

# Step 3: Interpretation of the results
print(f"Sample Mean: {sample_mean}")
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Step 4: Decision based on the p-value (alpha = 0.05)
alpha = 0.05  # Significance level

if p_value < alpha:
    print("\nSince the p-value is less than 0.05, we reject the null hypothesis.")
    print("There is a significant difference between the sample mean and the population mean.")
else:
    print("\nSince the p-value is greater than 0.05, we fail to reject the null hypothesis.")
    print("There is no significant difference between the sample mean and the population mean.")
```

```python
#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

import numpy as np
import scipy.stats as stats

# Step 1: Simulate random data from a normal distribution
np.random.seed(42)

# Parameters for the normal distribution
population_variance = 25  # Known population variance (sigma^2)
sample_size = 30          # Sample size
population_mean = 50      # Known population mean (for generating data)

# Simulate the sample data (e.g., sample from a population)
sample_data = np.random.normal(loc=population_mean, scale=np.sqrt(population_variance), size=sample_size)

# Step 2: Calculate sample variance
sample_variance = np.var(sample_data, ddof=1)  # ddof=1 to get the sample variance

# Step 3: Perform the Chi-square test for variance
# Degrees of freedom = sample size - 1
degrees_of_freedom = sample_size - 1

# Calculate the Chi-square test statistic
chi_square_stat = (sample_size - 1) * sample_variance / population_variance

# Step 4: Calculate the p-value for a two-tailed test
p_value = 2 * min(stats.chi2.cdf(chi_square_stat, degrees_of_freedom),
                  1 - stats.chi2.cdf(chi_square_stat, degrees_of_freedom))

# Step 5: Print results
print(f"Sample Variance: {sample_variance}")
print(f"Chi-square Statistic: {chi_square_stat}")
print(f"P-value: {p_value}")

# Step 6: Interpretation based on the p-value (alpha = 0.05)
alpha = 0.05  # Significance level

if p_value < alpha:
    print("\nSince the p-value is less than 0.05, we reject the null hypothesis.")
    print("There is a significant difference between the sample variance and the population variance.")
else:
    print("\nSince the p-value is greater than 0.05, we fail to reject the null hypothesis.")
    print("There is no significant difference between the sample variance and the population variance.")
```

```python
#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

import numpy as np
import scipy.stats as stats

# Sample data (number of successes and sample sizes)
# Group 1 (e.g., Treatment group)
x1 = 75  # number of successes in group 1
n1 = 100 # sample size for group 1

# Group 2 (e.g., Control group)
x2 = 60  # number of successes in group 2
n2 = 100 # sample size for group 2

# Step 1: Calculate sample proportions
p1 = x1 / n1
p2 = x2 / n2

# Step 2: Calculate the pooled proportion
P = (x1 + x2) / (n1 + n2)

# Step 3: Calculate the standard error
SE = np.sqrt(P * (1 - P) * (1/n1 + 1/n2))

# Step 4: Calculate the Z-statistic
Z = (p1 - p2) / SE

# Step 5: Calculate the p-value (for a two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(Z)))

# Step 6: Print results
print(f"Proportion of successes in Group 1: {p1}")
print(f"Proportion of successes in Group 2: {p2}")
print(f"Z-statistic: {Z}")
print(f"P-value: {p_value}")

# Step 7: Decision based on the p-value (alpha = 0.05)
alpha = 0.05  # Significance level

if p_value < alpha:
    print("\nSince the p-value is less than 0.05, we reject the null hypothesis.")
    print("There is a significant difference between the proportions of the two groups.")
else:
    print("\nSince the p-value is greater than 0.05, we fail to reject the null hypothesis.")
    print("There is no significant difference between the proportions of the two groups.")
```

```python
#26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate random data from normal distributions
np.random.seed(42)

# Parameters for the normal distributions (means and variances)
mean1 = 50
std1 = 10  # Standard deviation for dataset 1
std2 = 15  # Standard deviation for dataset 2

n1 = 30  # Sample size for dataset 1
n2 = 30  # Sample size for dataset 2

# Generate the data for both groups
data1 = np.random.normal(loc=mean1, scale=std1, size=n1)
data2 = np.random.normal(loc=mean1, scale=std2, size=n2)

# Step 2: Calculate the sample variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Step 3: Calculate the F-statistic
F_statistic = var1 / var2

# Step 4: Calculate the p-value (two-tailed test)
# Degrees of freedom: n1 - 1 for the numerator, n2 - 1 for the denominator
dof1 = n1 - 1
dof2 = n2 - 1

# Calculate the p-value for the F-distribution
p_value = 2 * min(stats.f.cdf(F_statistic, dof1, dof2),
                  1 - stats.f.cdf(F_statistic, dof1, dof2))

# Step 5: Print results
print(f"Variance of dataset 1: {var1}")
print(f"Variance of dataset 2: {var2}")
print(f"F-statistic: {F_statistic}")
print(f"P-value: {p_value}")

# Step 6: Decision based on the p-value (alpha = 0.05)
alpha = 0.05  # Significance level

if p_value < alpha:
    print("\nSince the p-value is less than 0.05, we reject the null hypothesis.")
    print("There is a significant difference between the variances of the two datasets.")
else:
    print("\nSince the p-value is greater than 0.05, we fail to reject the null hypothesis.")
    print("There is no significant difference between the variances of the two datasets.")

# Step 7: Visualizing the F-distribution and test statistic
x = np.linspace(0, 5, 1000)
y = stats.f.pdf(x, dof1, dof2)

plt.plot(x, y, label=f'F-distribution (df1={dof1}, df2={dof2})')
plt.axvline(F_statistic, color='r', linestyle='--', label=f'F-statistic = {F_statistic:.2f}')
plt.title('F-distribution for Variance Comparison')
plt.xlabel('F-statistic')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate data (e.g., roll of a fair die 100 times)
np.random.seed(42)

# Simulated data: Results from 100 rolls of a fair 6-sided die
observed = np.random.choice([1, 2, 3, 4, 5, 6], size=100)

# Step 2: Calculate observed frequencies
observed_freq = [np.sum(observed == i) for i in range(1, 7)]
print("Observed Frequencies:", observed_freq)

# Step 3: Define the expected frequencies (since the die is fair, each outcome should occur with probability 1/6)
expected_freq = [100 / 6] * 6  # Expected frequencies for a fair die

print("Expected Frequencies:", expected_freq)

# Step 4: Perform the Chi-square goodness of fit test
chi2_statistic, p_value = stats.chisquare(observed_freq, expected_freq)

# Step 5: Print the results
print(f"\nChi-square Statistic: {chi2_statistic}")
print(f"P-value: {p_value}")

# Step 6: Decision based on the p-value (alpha = 0.05)
alpha = 0.05  # Significance level

if p_value < alpha:
    print("\nSince the p-value is less than 0.05, we reject the null hypothesis.")
    print("There is a significant difference between the observed and expected frequencies.")
else:
    print("\nSince the p-value is greater than 0.05, we fail to reject the null hypothesis.")
    print("There is no significant difference between the observed and expected frequencies.")

# Step 7: Visualize the observed vs expected frequencies
categories = [1, 2, 3, 4, 5, 6]

plt.bar(categories, observed_freq, label='Observed Frequencies', alpha=0.6, color='blue')
plt.plot(categories, expected_freq, label='Expected Frequencies', linestyle='--', marker='o', color='red')
plt.xlabel('Die Faces')
plt.ylabel('Frequency')
plt.title('Chi-square Goodness of Fit: Observed vs Expected Frequencies')
plt.legend()
plt.show()
```
